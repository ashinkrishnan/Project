### Feature Engineering

###### Lets treat numeric and Categorical columns seperately also some columns contain numeric values mixed up with object we need to carefully extract those features too.

```python
# dropping unnamed column from both dataset
df_train = df_train.drop(['Unnamed: 0','New_Price'],axis=1)
df_test = df_test.drop(['Unnamed: 0','New_Price'],axis=1)
```

#### First of all We can change object columns into numeric that ares actually supposed to be numeric

```python
numset = ['Mileage','Engine','Power']
dftrain_numset = df_train[numset]
dftest_numset = df_test[numset]
```

```python
df_train[numset].columns
```

```python
df_test[['Mileage','addition1']] = df_test['Mileage'].str.split('.' , expand = True)
df_test[['Engine','addition2']] = df_test['Engine'].str.split(' ' , expand = True)
df_test[['Power','addition3']] = df_test['Power'].str.split('.' , expand = True)
df_test = df_test.drop(['addition1','addition2','addition3'],axis=1)
df_test.sample(5)
```

```python
def trans(dfcols):
    for col in dfcols.columns:
        dfcols[col] = pd.to_numeric(dfcols[col],errors='coerce')
        
    return dfcols
```

```python
df_test[numset] = trans(df_test[numset])
df_test.Mileage = df_test.Mileage.astype(float)
df_test
```

```python
df_train[['Mileage','addition1']] = df_train['Mileage'].str.split('.' , expand = True)
df_train[['Engine','addition2']] = df_train['Engine'].str.split(' ' , expand = True)
df_train[['Power','addition3']] = df_train['Power'].str.split('.' , expand = True)
df_train = df_train.drop(['addition1','addition2','addition3'],axis=1)
df_train.head(5)
```

```python
df_train[numset] = trans(df_train[numset])
df_train.Mileage = df_train.Mileage.astype(float)
df_train.head(5)
```
